## 🧹 Section 3 — Data Cleaning
> Handle missing values, fix column names, and remove duplicate/redundant columns.

### 3.1 — Check for Null Values
> Identify which columns have missing data.

```python
#Checking if there is null value.
df.isnull().sum()
```

### 3.2 — Fill Missing Values
> Fill missing `Review Rating` values with the median rating of the same category.  
> This preserves category-level patterns instead of using a global average.

```python
#. Filling the null value.
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

### 3.3 — Verify Null Values Are Fixed
> Confirm that no missing values remain after filling.

```python
#. Re-Checking if the null value if still exists.
df.isnull().sum()
```

### 3.4 — Standardize Column Names
> Convert all column names to lowercase and replace spaces with underscores for easier coding.

```python
#. For easy operation, changing all cases of top column in lower-case and replacing space of top column with '-'.
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

### 3.5 — Verify Column Names

```python
#. Checking the result
df.columns
```

### 3.6 — Rename Purchase Amount Column
> Simplify `purchase_amount_(usd)` to `purchase_amount` for cleaner query writing.

```python
#. Changing the name of column for easier operation.
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

### 3.7 — Verify Rename

```python
#. checking ..
df.columns
```

## ⚙️ Section 4 — Feature Engineering
> Create new columns that add business value and support better analysis.

### 4.1 — Create Age Group Column
> Segment customers into 4 age groups: `young_adult`, `adult`, `middle_aged`, `senior`  
> using quartile-based binning (`pd.qcut`).

```python
#Creating the age_group Column.
labels = ['young_adult', 'adult', 'middle_aged', 'senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

### 4.2 — Verify Age Group Column

```python
#. Result...
df[['age', 'age_group']].head(10)
```

### 4.3 — Create Purchase Frequency (Days) Column
> Convert text-based frequency (e.g. 'Weekly', 'Monthly') into numeric days for analysis.

```python
#. Create purchase_frequency_days column.
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

### 4.4 — Verify Purchase Frequency Column

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

### 4.5 — Check if Discount and Promo Code are the Same
> Verify whether `discount_applied` and `promo_code_used` carry identical information.

```python
#checking if all the discount given is throught promo-code.
(df['discount_applied'] == df['promo_code_used']).all()
```

### 4.6 — Drop Redundant Column
> Since `discount_applied` and `promo_code_used` are identical, drop `promo_code_used` to avoid duplication.

```python
# Since all the discount given is given through promo code, both columns are same .so,dropping one column.
df = df.drop('promo_code_used', axis = 1 )
```

### 4.7 — Final Column List
> Confirm the final set of columns after all cleaning and feature engineering.

```python
df.columns
```
